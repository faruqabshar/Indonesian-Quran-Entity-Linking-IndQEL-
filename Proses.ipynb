Excel Compare

```python
import pandas as pd
import numpy as np
df = pd.read_excel('ujicoba.xlsx')

condition = [
    (df['Rosyi'] == '_') & (df['Azmi'] == '_'),
    (df['Rosyi'] != '_') & (df['Azmi'] != '_') & (df['Rosyi'] == df['Azmi']),
    (df['Rosyi'] != '_') & (df['Azmi'] == '_'),
    (df['Rosyi'] == '_') & (df['Azmi'] != '_'),
    (df['Rosyi'] != '_') & (df['Azmi'] != '_') & (df['Rosyi'] != df['Azmi'])
]
result = ['Other', 'Agreed', 'OneNoLink','OneNoLink','Disagreed']
df['Diff'] = np.select(condition, result)
df
from pandas import ExcelWriter
writer = ExcelWriter('Hasil.xlsx')
df.to_excel(writer,'Sheet1',index=False)
writer.save()
```
